# Flight Log Notebook

This notebook creates battery and Rx voltage charts from a Spektrum tm1000 flight logger (a python replacement for flightlog.R).  The TLM files first need to be converted to CSV format using flighlog.py.

## Using flightlog.py

Check the inbuilt help with ``flightlog.py --help`` for any changes.  There are a couple of ways to reference the TLM files, either explicitly with the ``--tlm`` option or specifying the directory where the TLM files are stored using ``--dir`` and interactively selecting which file you want processed.  The second method is easier for casual use.

These notes are for Windows, the script should work with Linux but I haven't tried it.

1. Assumes you have a command line window open in the directory where ``flightlog.py`` is located.
1. If this is the first time you have run it, create the logs directory ``mkdir logs``.
1. Plug in the SD card from your Spektrum TX.  For this example I will assume the directory Windows assigns to the SD card is J:.
1. Run the program with ``flightlog.py --dir J:``.  You should be presented with an enumerated list of TLM files.
1. Hit the number for the file you want to process.  After this is run, the logs directory contain a copy of the TLM file, and also the processed CSV output.
1. Inspect the logs directory for the files with today's timestamp.

The newly created CSV file is the input to this notebook.

```python
%pylab inline
import pandas as pd
import numpy as np
```

```python
filename = './logs/1408031156-450xTgye.csv'
flog = pd.read_csv(filename)
flog.head()
```

We are only realy interested in the data rows.  If you have more than one heli with telemetry then you will probably want the header|flight start|model name|my model.

```python
# Header rows would be handy for further functionality but not 
# needed for my purposes.
flog = flog[flog.rectype == 'data']
#print(flog.shape)
flog = flog.drop(['offset', 'datatype'], axis=1)
# all value should now be numeric.
flog.value = flog.value.astype(np.float)
```

The records are in stored in a parameter/value type format which needs to be pivoted.  Not all values are sampled on each timestamp so the time series have embedded NaNs which are masked out at plot time.

```python
flightrecords = pd.pivot_table(flog, values='value', 
                               rows=['modelname', 'flightno', 'timestamp'], 
                               cols='parameter')
# Later on we join on this column, there is probably a less convoluted approach
flightrecords['flightno'] = flightrecords.index.get_level_values('flightno').values
flightrecords['timestamp'] = flightrecords.index.get_level_values('timestamp').values
print(flightrecords.head())
```

## Flight defragmentation

The flights can be fragmented depending on how the Tx is setup, ie with 4 batteries you would expect 4 flights however you will typically see more than 4.

We need to first extract the elapsed time from the fragments and treat them as a single flight if the gap is below some arbitrary threshold.  Check the timer settings for the model to see what conditions the timer is counting down.  If it is throttle > say 20% then in 3D flight you might keep breaking the flight (ie incrementing the flightno) when inverting.

First create a meta data frame, one record per flight and calculate new, combined flight numbers based on gap.

```python
# Create flight meta data data frame.
flightnos = np.unique(flightrecords.index.get_level_values('flightno').values)
flighttimes = pd.DataFrame([(fn, flightrecords.loc['450x Hitec', fn].index[0], 
           flightrecords.loc['450x Hitec', fn].index[-1])
                 for fn in flightnos],
                           columns=['flightno', 'start', 'end'])
mingap = 5
minelapsed = 10
# Add elapsed time for flightno
flighttimes['total'] = flighttimes.end - flighttimes.start
# Add gap between end of flight and start of next
flighttimes['gap'] = flighttimes.start - flighttimes.end.shift()
# If thie gap is long enough, flag it as a new flight.
flighttimes['newflag'] = 1 * (flighttimes.gap > mingap)
# Relabel genuine flights with vflightno
flighttimes['vflightno'] = flighttimes.newflag.cumsum()
print(flighttimes)
```

Join the newly calculated flightnos back to the original data frame.

```python
flightrecords = flightrecords.merge(flighttimes, on='flightno')
flightrecords.set_index(['vflightno'], inplace=True)
print(flightrecords.head())
```

```python
vflightnos = flightrecords.index.get_level_values('vflightno').unique()
print(vflightnos)
```
